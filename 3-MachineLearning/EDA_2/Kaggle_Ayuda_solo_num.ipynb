```python
import re

def separate_screen_resolution(resolution_str):
    # Extract resolution
    resolution_match = re.search(r'(\d+)\s*x\s*(\d+)', str(resolution_str))
    resolution = f"{resolution_match.group(1)}x{resolution_match.group(2)}" if resolution_match else None
    
    # Extract screen type
    screen_types = ['Touchscreen', 'IPS Panel', 'Full HD', 'Quad HD+', '4K Ultra HD', 'Retina Display']
    found_types = [type for type in screen_types if type.lower() in resolution_str.lower()]
    screen_type = ' / '.join(found_types) if found_types else None
    
    return screen_type, resolution

# Apply the function
test[['ScreenType', 'Resolution']] = test['ScreenResolution'].apply(separate_screen_resolution).apply(pd.Series)
```

```python
# Crear una nueva columna 'IsTouchscreen' con True si contiene 'Touchscreen' y False si no
test['IsTouchscreen'] = test['ScreenType'].str.contains('Touchscreen', case=False, na=False)
```

```python
# Separar el tipo de CPU y la memoria (en GHz)
test[['Cpu_Type', 'Memory_Cpu']] = test['Cpu'].str.extract(r'^(.*)\s(\d+\.?\d*GHz)$')

test.head()
```

```python
# Función para limpiar los valores y normalizarlos
def clean_and_normalize_memory(value):
    # Asegurarnos de que el valor es un string
    value = str(value).lower().strip()
    
    # Usar una expresión regular para capturar el número y la unidad opcional
    match = re.match(r'(\d+(\.\d+)?)(\s*(gb|mb|tb)?)?', value)
    
    if match:
        # Extraer el número y la unidad
        number = float(match.group(1))  # Captura el número
        unit = match.group(3).strip() if match.group(3) else ''  # Captura la unidad (puede estar vacía)
        
        # Convertir todo a GB
        if 'gb' in unit or unit == '':
            return number  # Ya está en GB o es solo un número
        elif 'mb' in unit:
            return number / 1024  # Convertir de MB a GB
        elif 'tb' in unit:
            return number * 1024  # Convertir de TB a GB
    return None  # Si no es válido, devolver None

# Aplicar la función a la columna Memory
test['Memory'] = test['Memory'].apply(clean_and_normalize_memory)
```

```python
test[['OpsysType', 'Version']] = test['OpSys'].str.extract(r'^(Windows|Chrome OS|Linux|No OS|Mac OS X|macOS|Android)(?:\s+(.*))?$')

test.head()
```

```python
test[['Manufacturer', 'MemoryGraphic']] = test['Gpu'].str.extract(r'^(Intel|AMD|Nvidia|ARM)\s+([\w\s]+)')

test.head()
```

```python
test['Model Base'] = test['Product'].str.extract(r'^([^(/]+)')  # Extrae el texto antes de un paréntesis o diagonal
test['Specifications'] = test['Product'].str.extract(r'\((.*)\)|/(.*)').bfill(axis=1)[0]  # Extrae el contenido entre paréntesis o después de una diagonal

test.head()
```

```python
test['Ram'] = test['Ram'].astype(str)
test['Ram'] = test['Ram'].str.replace('GB', '').astype(float)
test['Memory'] = test['Memory'].astype(str)
test['Memory'] = test['Memory'].str.replace('GHz', '').astype(float)
test['Weight']= test['Weight'].astype(str)
test['Weight'] = test['Weight'].str.replace('kg', '').astype(float)
test['Memory_Cpu']= test['Memory_Cpu'].astype(str)
test['Memory_Cpu'] = test['Memory_Cpu'].str.replace('GHz', '').astype(float)
```
