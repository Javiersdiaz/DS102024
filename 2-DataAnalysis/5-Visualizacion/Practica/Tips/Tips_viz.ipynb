### Paso 10. Cree dos histogramas del valor del tip basados en Dinner y Lunch. Deben estar uno al lado del otro.

```python
# Crear la figura y los subgr√°ficos
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Histograma para Dinner
sns.histplot(data=tips[tips["time"] == "Dinner"], x="tip", bins=20, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Distribuci√≥n de Tip (Dinner)")
axes[0].set_xlabel("Tip")
axes[0].set_ylabel("Frecuencia")

# Histograma para Lunch
sns.histplot(data=tips[tips["time"] == "Lunch"], x="tip", bins=20, kde=True, color="green", ax=axes[1])
axes[1].set_title("Distribuci√≥n de Tip (Lunch)")
axes[1].set_xlabel("Tip")
axes[1].set_ylabel("Frecuencia")

# Ajustar espacio entre los subgr√°ficos
plt.tight_layout()
plt.show()
```

### Paso 11. Cree dos gr√°ficos de dispersi√≥n, uno para los hombres y otro para las mujeres, que presenten la relaci√≥n entre el valor ``total_bill`` y la ``tip``, diferenciando entre ``smoker`` y ``no smoker``.
### Deben estar uno al lado del otro.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Gr√°fico de dispersi√≥n para hombres
sns.scatterplot(
    data=tips[tips["sex"] == "Male"],
    x="total_bill", y="tip", hue="smoker",
    palette="coolwarm", style="smoker", ax=axes[0]
)
axes[0].set_title("Hombres: Relaci√≥n Total Bill vs Tip")
axes[0].set_xlabel("Total Bill")
axes[0].set_ylabel("Tip")
axes[0].legend(title="Fumador")

# Gr√°fico de dispersi√≥n para mujeres
sns.scatterplot(
    data=tips[tips["sex"] == "Female"],
    x="total_bill", y="tip", hue="smoker",
    palette="coolwarm", style="smoker", ax=axes[1]
)
axes[1].set_title("Mujeres: Relaci√≥n Total Bill vs Tip")
axes[1].set_xlabel("Total Bill")
axes[1].set_ylabel("Tip")
axes[1].legend(title="Fumador")

# Ajustar espacio entre los subgr√°ficos
plt.tight_layout()
plt.show()
```
